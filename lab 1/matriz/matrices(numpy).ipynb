# Multiplicación de matrices (numpy)

se importan las librerías que se usan

```python
import time
import matplotlib.pyplot as plt
import numpy as np
```

se realiza la multiplicación para varios n

```python
n = 200
ns = []
tiempos = []
while (n <= 1000):
    suma = 0
    i = 0
    ns.append(n)
    a = np.ones((n,n))
    b = np.full((n, n), 2)
    while(i < 10 ):
        tiempo_inicial = time.time()
        (a*b)
        tiempo_final = time.time() - tiempo_inicial
        suma = (suma + tiempo_final)
        i += 1
    tiempos.append(suma/10)
    n +=200

```

Se grafica

```python
plt.plot(ns,tiempos)
plt.xlabel("n");
plt.ylabel("Tiempo");
```
